```python
import pandas as pd
import re

from numpy import number

splits = {'train': 'main/train-00000-of-00001.parquet', 'test': 'main/test-00000-of-00001.parquet'}
df = pd.read_parquet("hf://datasets/openai/gsm8k/" + splits["train"])
df.columns
```

```python
number_re = re.compile(r'####.(\d+)')
string = """If Washington has 140 goats, Washington has 140+40 = <<140+40=180>>180 goats. In total, they have 140+180 = <<140+180=320>>320 goats #### 320"""
string_re = number_re.findall(string)
number_re = int(string_re[0])
number_re
```

```python
def transform_to_number(string_series: pd.Series):
    number_re = re.compile(r'####.(-?\d+)')
    res = []
    for string in string_series.values:
        string_re = number_re.findall(string)
        if string_re:
            number = int(string_re[0])
            res.append(number)
        else:
            raise ValueError(f"No number found, string: {string}")
    return res
transform_to_number(df['answer'])
```
